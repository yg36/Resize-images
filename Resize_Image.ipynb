```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

img=cv2.imread("colors.webp")
print(img.shape)
cv2_imshow(img)
cv2.waitKey(0)
```

```python
# resizing the images
img_resize=cv2.resize(img, (256,256))
cv2_imshow(img_resize)
cv2.waitKey(0)
print(img_resize.shape)
```
